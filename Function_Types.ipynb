```python
"""
Function types:

> positional arguments
> default arguments
> keyword arguments
> variable arguments

"""
```

```python

def add(a,b,c,d):
  result = a+b+c+d
  return result

add(1,5, 6,6)
```

```python
def prod(x1, x2, x3=1): # default arguments are always last in order
  result = x1*x2*x3
  return result

prod(1,2,3)
```

```python
def greet (usr, message):
  result = "Hello" +usr+ "your message is -->" + message
  return result

greet("vijay", "Good Morning!")
```

```python
greet("Good Morning!", "vijay")
```

```python
greet(message="Good Morning!", usr="vijay")
```

```python
def prod(*value):
  print(value, type(value))
  result = 1
  for i in value:
    result = result*i
  return result

prod(1,3,4,5,6,7,8,9,4,57,8,9,9,9,8)
```
