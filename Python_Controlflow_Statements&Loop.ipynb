```python
'''
Control Structures or Control Flow Statements

Decisions making statemtents : if, if-else, elif, match

Repetitive or iterative or looping statements: while, for

if:

  If condition:

    statements
  if conditon:
    statements
  else: --blind
    statements

  if condition:
    statements
  elif condition:
    statements
  elif condition:
    statements
  else: #optional
    statements

  while:
    while condition:
      statements


'''
```

```python
def main():
  x = eval(input('enter your first number:'))
  y = eval(input('enter your second number:'))
  if x > y:
    print(x+y)
  elif x <y:
    print(x-y)
  elif x == y:
    print(x*y)
  else:
    print('I can not perform any operation:')
main()
```

```python
'''
while:
    while condition:
      statements

  #1- intial value: from where to start
  #2- terminating condition: where loop should stop
  #3 - step value: how to loop has to navigate

Note: while loop repeats the statements until the condition is true even if one time condition is false then loop will terminate.

for:

  for variable in iterable
    statements

  iterable: list, tuple, sets, dict etc.

'''
```

```python
def main():
  i = 1  #initial
  while i <=8:  #terminating
    print(i) #print value
    i = i + 1 #step value
main()
```

```python
"""
Match:

Case statements in SQL. Case --> when --> then --> else

match expression:
  case value1:
    statements
  case value2:
    statements
  case value3:
    statements
  case _:
    statements

"""
```

```python
#Weeksdays
day =6
match day: #6
  case 1:
    print('Monday')
  case 2:
    print('Tuesday')
  case 3:
    print('Wednesday')
  case 4:
    print('Thursday')
  case 5:
    print('Friday')
  case _:
    print('no match')
```

```python
day =7
match day:
  case 1 | 2 | 3 | 4 | 5:
    print('Weekday')
  case 6 | 7:
    print('Weekend')
  case _:
    print('no match')
```

```python
day =4
month=5
match day:
  case 1 | 2 | 3 | 4 | 5 if month ==4:
    print('April')
  case 1 | 2 | 3 | 4|5 if month == 5:
    print('May')
  case _:
    print('no match')
```

```python
def main():
  i = 0  #initial
  while i <=5:  #terminating
    if i == 5: #true 3 ==3
      break
    print(i) #print value
    i = i + 1 #step value
main()
```

```python
def main():
  i=0
  while i <=5:  #terminating
    i = i +1
    if i ==5:
      continue
    print(i)
main()
```
